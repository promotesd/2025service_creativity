#### 概率论与数理统计(浙大四版).pdf

```python
from pdf2image import convert_from_path
import pytesseract
import os

def pdf_to_text_single_file(pdf_path, output_file, lang='chi_sim'):
    """
    将 PDF 文件转换为文本，并将所有页面的 OCR 结果依次追加到同一个文件中。
    
    参数:
    - pdf_path: PDF 文件路径
    - output_file: 输出的文本文件路径（所有结果都写到这里）
    - lang: OCR 语言包（例如：'eng' 英文，'chi_sim' 简体中文，'chi_tra' 繁体中文）
    """
    
    # 如果输出文件已存在，建议先清空或者备份，以免混淆
    # 如果你需要保留原有内容，可以不执行下面这行
    open(output_file, 'w', encoding='utf-8').close()
    
    # 将 PDF 转换为图像列表，每页一张图片（默认分辨率为 300）
    pages = convert_from_path(pdf_path, dpi=300)
    
    for i, page in enumerate(pages):
        # 对每一页图像进行 OCR 识别
        text = pytesseract.image_to_string(page, lang=lang)
        
        # 可以在每页内容前后做一些分割行，方便阅读
        page_text = f"\n\n--- 第 {i+1} 页 ---\n{text}\n"
        
        # 在控制台打印当前页的识别结果
        print(page_text)
        
        # 将当前页的结果追加写入到 output_file
        with open(output_file, 'a', encoding='utf-8') as f:
            f.write(page_text)

    print("OCR 处理完成！全部内容已写入：", output_file)


if __name__ == '__main__':
    # PDF 文件路径
    pdf_file = r'/root/autodl-tmp/dataset/mathbook/大学数学/概率论与数理统计(浙大四版).pdf'
    
    # 指定的输出文本文件，如果文件存在可以被覆盖或追加
    output_file = r'/root/autodl-tmp/code/2025service_creativity/process_dataset/OCRmathbook/概率论与数理统计(浙大四版).txt'
    
    # 如果是中文文本，请使用 'chi_sim'（简体中文）或 'chi_tra'（繁体中文）
    pdf_to_text_single_file(pdf_file, output_file, lang='chi_sim')
    print("OCR 处理完成！")
```
